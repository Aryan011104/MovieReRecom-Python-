```python
import pandas as pd
data=pd.read_csv("imdb_top_1000.csv")
data.head()
```

```python
data.columns
```

```python
data['Certificate'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
rat=LabelEncoder()
rat.fit(data['Certificate'])
na=LabelEncoder()
na.fit(data['Series_Title'])
ti=LabelEncoder()
ti.fit(data['Runtime'])
```

```python
#1st GPT: How to import train test split
#2nd GPT: Removing brackets from values()
from sklearn.model_selection import train_test_split
data['Certificate1']=rat.transform(data['Certificate'])
data['name']=na.transform(data['Series_Title'])
data['time']=ti.transform(data['Runtime'])
X=data[['IMDB_Rating','time','Certificate1']].values
y=data['name'].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
import numpy as np
certification=input("Enter Certification of the movie")
rating=float(input("Enter the rating of the movie"))
time=input("Enter the length of the movie")
certification1=rat.transform([certification])[0]
time1=ti.transform([time])[0]
z=np.array([[rating,time1,certification1]])
output=model.predict(z)
output1=na.inverse_transform(output)
print(output1)
movie=output1[0]
description=data.loc[data['Series_Title']==movie,"Overview"].values
print(description)
```
